```python
from scipy.stats import zscore
z=np.abs(zscore(Data))
z
```

```python
print(np.where(z>3))
```

```python
Data_new=Data[(z<3).all(axis=1)]
Data_new
```

```python
print(Data.shape)
print(Data_new.shape)
```

```python
Data_new.skew()#After Removing the outliers
```

```python
from scipy.stats import boxcox
Data_new['MedInc']=boxcox(Data_new['MedInc'],0)
```

```python
list1=['AveRooms','AveBedrms','Population','AveOccup','House_price']
for i in list1:
    Data_new[i]=boxcox(Data_new[i],0)
```

```python
Data_new.skew()
```

```python
sns.pairplot(Data_new)
```

```python
corr=Data_new.corr()
plt.figure(figsize=(12,12))
sns.heatmap(corr,annot=True)
```

```python
Data_new.drop(['Longitude'],axis=1,inplace=True)
```

```python
Data_new
```

```python
x=Data_new.iloc[:,0:-1]
x.head()
```

```python
y=Data_new.iloc[:,-1].values.reshape(-1,1)
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import cross_val_score
max_r2_score=0
for r_state in range(42,101):
    train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=r_state,test_size=0.40)
    lr=LinearRegression()
    lr.fit(train_x,train_y)
    pred=lr.predict(test_x)
    r2_scr=r2_score(pred,test_y)
    if r2_scr>max_r2_score:
        max_r2_score=r2_scr
        final_r_state=r_state
    print('Max r2 score corresponding to ',final_r_state,' is ',max_r2_score)    
```

```python
a_score=cross_val_score(lr,x,y,cv=10)
print(a_score)
print(a_score.mean())
print(np.std(a_score))
```

```python
#Final LinearRegression
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=100,test_size=0.40)
lr.fit(train_x,train_y)
lrpred=lr.predict(test_x)
print('Predicted Value',lrpred)
print('Real Value',test_y)
```

```python
print('R2 score',r2_score(lrpred,test_y))
print('Mean Absolute Error',mean_absolute_error(lrpred,test_y))
print('Mean Squared Error',mean_squared_error(lrpred,test_y))
print('Root Mean Squared Error',np.sqrt(mean_squared_error(lrpred,test_y)))
```

```python
from sklearn.svm import SVR
sv=SVR()
sv.fit(train_x,train_y)
svpred=sv.predict(test_x)
print('Predicted Value',svpred)
print('Real Value',test_y)
```

```python
print('R2 score',r2_score(svpred,test_y))
print('Mean Absolute Error',mean_absolute_error(svpred,test_y))
print('Mean Squared Error',mean_squared_error(svpred,test_y))
print('Root Mean Squared Error',np.sqrt(mean_squared_error(svpred,test_y)))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(train_x,train_y)
dtpred=dt.predict(test_x)
print('Predicted Value',dtpred)
print('Real Value',test_y)
```

```python
print('R2 score',r2_score(dtpred,test_y))
print('Mean Absolute Error',mean_absolute_error(dtpred,test_y))
print('Mean Squared Error',mean_squared_error(dtpred,test_y))
print('Root Mean Squared Error',np.sqrt(mean_squared_error(dtpred,test_y)))
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=10)
knn.fit(train_x,train_y)
knpred=knn.predict(test_x)
print('Predicted Value',knpred)
print('Real Value',test_y)
```

```python
print('R2 score',r2_score(knpred,test_y))
print('Mean Absolute Error',mean_absolute_error(knpred,test_y))
print('Mean Squared Error',mean_squared_error(knpred,test_y))
print('Root Mean Squared Error',np.sqrt(mean_squared_error(knpred,test_y)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(train_x,train_y)
rfpred=rf.predict(test_x)
print('Predicted Value',rfpred)
print('Real Value',test_y)
```

```python
print('R2 score',r2_score(rfpred,test_y))
print('Mean Absolute Error',mean_absolute_error(rfpred,test_y))
print('Mean Squared Error',mean_squared_error(rfpred,test_y))
print('Root Mean Squared Error',np.sqrt(mean_squared_error(rfpred,test_y)))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gd=GradientBoostingRegressor()
gd.fit(train_x,train_y)
gdpred=gd.predict(test_x)
print('Predicted Value',gdpred)
print('Real Value',test_y)
```
