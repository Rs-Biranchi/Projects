# Machine Learning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score,train_test_split
max_acc_scr=0
for r_state in range(42,101):
    train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=r_state,test_size=0.30)
    lg=LogisticRegression()
    lg.fit(train_x,train_y)
    pred=lg.predict(test_x)
    acc_scr=accuracy_score(pred,test_y)
    if acc_scr>max_acc_scr:
        max_acc_scr=acc_scr
        final_r_state=r_state
    print('Max accuracy score corresponding to ',final_r_state,' is ',max_acc_scr)    
```

```python
#Checking if the above model is overfitting or underfitting
a_score=cross_val_score(lg,x,y,cv=10)
print(a_score)
print(np.mean(a_score))
```

```python
#Final LogisticRegression
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=93,test_size=0.3)
lg=LogisticRegression()
lg.fit(train_x,train_y)
lgpred=lg.predict(test_x)
print('Predicted Attrition ',lgpred)
print('Actual Attrition ',test_y)
```

```python
print('Accuracy Score',accuracy_score(lgpred,test_y))
print('Confusion Matrix \n',confusion_matrix(lgpred,test_y))
print('Classification Report \n',classification_report(lgpred,test_y))
```

Performing  Anothers models

```python
from sklearn.svm import SVC
def svkernel(ker):
    sv=SVC(kernel=ker)
    sv.fit(train_x,train_y)
    svpred=sv.predict(test_x)
    print('Predicted Attrition',svpred)
    print('\n')
    a_scr=cross_val_score(sv,x,y)
    print(' cross val scr',a_scr)
    print('mean of Cross val scr',np.mean(a_scr))
    print('\n')
    print('Accuracy Score',accuracy_score(svpred,test_y))
    print('Confusion Matrix \n',confusion_matrix(svpred,test_y))
    print('Classification Report \n',classification_report(svpred,test_y))
```

```python
from sklearn.tree import DecisionTreeClassifier
def dtcriterion(cri):
    dt=DecisionTreeClassifier(criterion=cri)
    dt.fit(train_x,train_y)
    dtpred=dt.predict(test_x)
    print('Predicted Attrition',dtpred)
    print('\n')
    a_scr=cross_val_score(dt,x,y)
    print(' cross val scr',a_scr)
    print('mean of Cross val scr',np.mean(a_scr))
    print('\n')
    print('Accuracy Score',accuracy_score(dtpred,test_y))
    print('Confusion Matrix \n',confusion_matrix(dtpred,test_y))
    print('Classification Report \n',classification_report(dtpred,test_y))
```

```python
from sklearn.neighbors import KNeighborsClassifier
def knneighbor(n):
    kn=KNeighborsClassifier(n_neighbors=n)
    kn.fit(train_x,train_y)
    knpred=kn.predict(test_x)
    print('Predicted Attrition',knpred)
    print('\n')
    a_scr=cross_val_score(kn,x,y)
    print(' cross val scr',a_scr)
    print('mean of Cross val scr',np.mean(a_scr))
    print('\n')
    print('Accuracy Score',accuracy_score(knpred,test_y))
    print('Confusion Matrix \n',confusion_matrix(knpred,test_y))
    print('Classification Report \n',classification_report(knpred,test_y))
```
